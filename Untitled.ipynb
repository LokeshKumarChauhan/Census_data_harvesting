```python
for index, row in df.iterrows():
    insert_query = '''
    INSERT INTO customers (ID, Name, City, State, ZIP)
    VALUES (%s, %s, %s, %s, %s)
    '''
    cursor.execute(insert_query, (row[0], row[1], row[2], row[3], row[4]))  # Using indices for each column

# Commit the transaction
conn.commit()

print("Data inserted successfully!")
```
